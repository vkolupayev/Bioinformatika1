### Ta pati veiksmu seka kiekvienam likusiem fasta failams

```
def prep_func(file):
    for seq_record in SeqIO.parse(file, "fasta"):
        print(seq_record)
    seq_str = str(seq_record.seq)
    seq_str_revcompl = str(seq_record.reverse_complement().seq)
    seq_str_1rf = re.findall(r'ATG(?:(?!TAA|TAG|TGA)...)*(?:TAA|TAG|TGA)',seq_str)
    seq_str_2rf = re.findall(r'ATG(?:(?!TAA|TAG|TGA)...)*(?:TAA|TAG|TGA)',seq_str[1:-2])
    seq_str_3rf = re.findall(r'ATG(?:(?!TAA|TAG|TGA)...)*(?:TAA|TAG|TGA)',seq_str[2:-1])
    seq_str_revcompl_1rf =re.findall(r'ATG(?:(?!TAA|TAG|TGA)...)*(?:TAA|TAG|TGA)',seq_str_revcompl)
    seq_str_revcompl_2rf = re.findall(r'ATG(?:(?!TAA|TAG|TGA)...)*(?:TAA|TAG|TGA)',seq_str_revcompl[1:-2])
    seq_str_revcompl_3rf = re.findall(r'ATG(?:(?!TAA|TAG|TGA)...)*(?:TAA|TAG|TGA)',seq_str_revcompl[2:-1])
    seq_str_1rf_filtered= list(filter(over100 , seq_str_1rf))
    seq_str_2rf_filtered= list(filter(over100 , seq_str_2rf))
    seq_str_3rf_filtered= list(filter(over100 , seq_str_3rf))
    seq_str_revcompl_1rf_filtered= list(filter(over100 , seq_str_revcompl_1rf))
    seq_str_revcompl_2rf_filtered= list(filter(over100 , seq_str_revcompl_2rf))
    seq_str_revcompl_3rf_filtered= list(filter(over100 , seq_str_revcompl_3rf))

    codon_dicodon_dict = {}
    codon = [''.join(i) for i in product(['T', 'C', 'A', 'G'], repeat = 3)]
    dicodon = [''.join(i) for i in product(codon, repeat=2)]

    codon_dict = {c:0 for c in codon}
    dicodon_dict = {dc:0 for dc in dicodon}
    
    #merge both dictionaries python 3.5 >
    codon_dicodon_dict = {**codon_dict, **dicodon_dict}

    for protein in seq_str_1rf_filtered:
        for key in codon_dict:
            codon_dict[key] += protein.count(key)
        for key in dicodon_dict:
            dicodon_dict[key] += protein.count(key)
            
    for protein in seq_str_2rf_filtered:
        for key in codon_dict:
            codon_dict[key] += protein.count(key)
        for key in dicodon_dict:
            dicodon_dict[key] += protein.count(key)

    for protein in seq_str_2rf_filtered:
        for key in codon_dict:
            codon_dict[key] += protein.count(key)
        for key in dicodon_dict:
            dicodon_dict[key] += protein.count(key)

    #reverse complement
    for protein in seq_str_revcompl_1rf_filtered:
        for key in codon_dict:
            codon_dict[key] += protein.count(key)
        for key in dicodon_dict:
            dicodon_dict[key] += protein.count(key)
            
    for protein in seq_str_revcompl_2rf_filtered:
        for key in codon_dict:
            codon_dict[key] += protein.count(key)
        for key in dicodon_dict:
            dicodon_dict[key] += protein.count(key)

    for protein in seq_str_revcompl_2rf_filtered:
        for key in codon_dict:
            codon_dict[key] += protein.count(key)
        for key in dicodon_dict:
            dicodon_dict[key] += protein.count(key)

    cod_sum = 0
    dicod_sum = 0
    for key in codon_dict:
        cod_sum += codon_dict[key]
    for key in dicodon_dict:
        dicod_sum += dicodon_dict[key]

    for key in codon_dict:
        codon_dict[key] = codon_dict[key]/cod_sum
    for key in dicodon_dict:
        dicodon_dict[key] = dicodon_dict[key]/dicod_sum

    return codon_dict, dicodon_dict
```
